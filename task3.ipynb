## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img = plt.imread("data/astronaut.png")
plt.imshow(img, 'gray')
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
import numpy as np
def meanfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] >= n and p[0] < img_in.shape[0] and p[1] >= n and p[1] < img_in.shape[1]:
            for q in numpy.ndindex(size, size):
                neighborhood_sum = neighborhood_sum + img_in[p[0] + q[0] - 2*n, p[1] + q[1] - 2*n]
            new_pixel = neighborhood_sum/(size*size)
            result[p] = new_pixel
        else:
            result[p] = 0
    plt.imshow(result, 'gray')
```
